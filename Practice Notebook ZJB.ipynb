## 5.  Write down a function called "inspect" to inspect array1 and array2 Shape, length, Dimensions, and Datatype using Preetytable. Your pretty table should contain a title and look like similar to the following:


```python

def inspect(array):
    PT=PrettyTable()
    PT.title="Numpy array Inspection"
    PT.field_names=['Shape', 'len', 'Dimensions', 'Datatype']
    PT.add_row([array.shape, len(array), array.ndim, array.dtype])
    print(PT)    
```

```python
inspect(array1)
```

```python
inspect(array2)
```

## 8.Create a function called "calculation" to display `sum, min, max, mean, standard deviation, median calculations about a NumPy array using PrettyTable. Test it for array1,2,3


```python
def calculation(array):
    PT=PrettyTable()
    PT.title='Numpy array Calculations'
    PT.field_names=['sum','min','max','Mean','Std', 'median']
    PT.add_row([array.sum(),array.min(),array.max(),array.mean(),np.std(array),np.median(array)])
    print(PT)
```

```python
calculation(array1)
calculation(array2)
calculation(array3)
```

## 9.Create a function called "arithmatic" to display `Subtraction (np.subtract(a,b)), Addition,  Multiplication calculations about two NumPy array using PrettyTable. Test it for 
* array1, array3
* array2, array3
* array1, array2


```python
def arithmatic(a,b):
    PT=PrettyTable()
    PT.title="Numpy array Arithmatic Operations"
    PT.field_names=['Subtraction', 'Addition','Multiplication']
    PT.add_row([np.subtract(a,b),np.add(b,a),np.multiply(a,b)])
    print(PT)
```

```python
arithmatic(array1,array3)
```

```python
arithmatic(array2,array3)
```

```python
arithmatic(array1,array2)
```
